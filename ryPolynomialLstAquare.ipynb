# 多項式函數逼近隨機資料點

- 給定 N 組 x, y 的對應
- 找 1條多項式函數來逼近它。


```python
import numpy as np
import matplotlib.pyplot as pl


x= np.linspace(-10,10, 21)
x
```

```python
f= np.polynomial.Polynomial([1,-1,1,-1])
f
```

```python
y= f(x)
y
```

```python
z= np.random.normal(0,100,size=y.size)
z
```

```python
q= y+z
q
```

```python

pl.plot(x,y,'r',
        x,z,'g',
        x,q,'bx'
        )
pl.grid()
```

```python
pl.plot(#x,y,'r',
        #x,z,'g',
        x,q,'bx'
        )
pl.grid()
```

```python
x,q
```

```python
A= x[:,None]
A= A**[0,1]
A

a= np.linalg.lstsq(A,q)
a= a[0]
a

g= np.polynomial.Polynomial(a)
g

yHat= g(x)
pl.plot(x,y,'r',
        x,yHat,'g',
        x,q,'bx'
        )
pl.grid()
```

```python

A= x[:,None]
A= A**[0,1,2]
A

a= np.linalg.lstsq(A,q)
a= a[0]
a

g= np.polynomial.Polynomial(a)
g

yHat= g(x)
pl.plot(x,y,'r',
        x,yHat,'g',
        x,q,'bx'
        )
pl.grid()
```
